```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
from bokeh.models import HoverTool
from IPython.display import HTML,display
from sklearn.ensemble import IsolationForest
from sklearn import tree
import warnings
warnings.filterwarnings("ignore")
```

```python
#df= pd.read_excel('/content/drive/MyDrive/SC_TASK Forecast/sc_task.xlsx')
df= pd.read_csv('../data/CPN50_Preprocessed.csv', sep='|',index_col=0)
```

```python
df
```

```python
df = df.dropna(subset=['r1_iram1622'])
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba (80% entrenamiento, 20% prueba)
#rain_size = int(len(df) * 0.8)
#train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
# Visualización de distribución de datos
plt.figure(figsize=(18, 6))
sns.histplot(df["r1_iram1622"], bins=20, kde=True, color='blue')
plt.title('Value Distribution of Full Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Entrenar el modelo en todo el dataset
model_daily = IsolationForest(random_state=0, contamination=0.03)
model_daily.fit(df[['r1_iram1622']])
```

```python
# Detección de outliers en todo el dataset
outliers_daily = pd.Series(model_daily.predict(df[['r1_iram1622']])).apply(lambda x: 1 if x == -1 else 0).to_numpy()
```

```python
#Anomaly score
anomaly_score_daily = model_daily.decision_function(df[['r1_iram1622']])
```

```python
#add prediction results to the data frame for visualization purpose
df['outliers'] = outliers_daily
df['anomaly_score'] = anomaly_score_daily
df = df.reset_index()  # Restablecer el índice para visualización
```

```python
# Configurar las herramientas de visualización
tooltips = [
    ('index', '@index'),
    ('Value', '@r1_iram1622'),
    ('Outliers', '@outliers'),
    ('Anomaly_Score', '@anomaly_score')
]

hover = HoverTool(tooltips=tooltips)

# Visualizar los datos con anomalías y valores normales
hv.Points(df.query("outliers == 1"), ['index', 'r1_iram1622']).opts(size=8, color='#ff0000') * \
hv.Points(df.query("outliers == 0"), ['index', 'r1_iram1622']).opts(size=8, color='#048c2d') * \
hv.Curve(df, ['index', 'r1_iram1622']).opts(
    opts.Curve(
        title="strength day 1 - Anomalies and Normal Data",
        xlabel="Index",
        ylabel="Value",
        height=300,
        responsive=True,
        tools=[hover, 'box_select', 'lasso_select', 'tap'],
        show_grid=True
    )
)
```

```python
df = df[df['outliers'] != 1]
```

```python
# Eliminar las columnas 'outliers' y 'anomaly_score'
df = df.drop(['outliers', 'anomaly_score'], axis=1)
```

```python

# Define the size of the training set (80% of the total).
train_size = int(len(df) * 0.8)

# Split the dataset into training and test without mixing the temporal order.
df_train = df.iloc[:train_size]
df_test = df.iloc[train_size:]

df_train.to_csv("../data/train/D1training.csv", index=False)
df_test.to_csv("../data/test/D1test.csv", index=False)
```
